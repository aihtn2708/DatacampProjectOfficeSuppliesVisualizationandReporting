# **Analytics**

## **Region wise** / **category wise**


```python
region_wise=office_supplies.groupby(['Region'])[ 'Sales', 'Quantity', 'Profit'].sum()
sorted_region=region_wise.sort_values('Sales', ascending=False)
print(sorted_region)
```

```python
plt.rcParams['figure.figsize']=(10,10)
sorted_region.plot.barh()
```

```python
subcat_wise=office_supplies.groupby(['Sub-Category'])[ 'Sales', 'Quantity', 'Profit'].sum()
sorted_subcat=subcat_wise.sort_values('Sales', ascending=False)
print(sorted_subcat)
```

```python
plt.rcParams['figure.figsize']=(10,10)
sorted_subcat.plot.barh()
```

```python
stacked_data = pd.pivot_table(office_supplies,index=['Category'], columns=['Region'], values=['Sales'], aggfunc='sum')
stacked_data.plot.barh(stacked=True)
```

```python
mekko =  pd.pivot_table(office_supplies,index=['Region'], columns=['Category'], values=['Sales'], aggfunc='sum')

mekko.columns = mekko.columns.droplevel(0)
mekko = mekko.reset_index().rename_axis(None, axis=1)


bars1 = mekko['Furniture']
bars2 = mekko['Office Supplies']
bars3 = mekko['Technology']


bars = np.add(bars1, bars2).tolist()


r = mekko['Region']


plt.bar(r, bars1, color='r', edgecolor='white', width=1)

plt.bar(r, bars2, bottom=bars1, color='y', edgecolor='white', width=1)
# Create green bars (top)
plt.bar(r, bars3, bottom=bars, color='g', edgecolor='white', width=1)
plt.xticks(r, r, fontweight='bold');plt.xlabel("Region");plt.ylabel("Category");plt.legend(office_supplies.Category.unique(),loc=2)
plt.show()
```

```python
grouped=pd.DataFrame(office_supplies.groupby(['year','Ship Mode','Category','Sub-Category','Region'])['Quantity','Discount','Sales','Profit'].sum().reset_index())
grouped
```

```python
sns.set_theme(style="dark")
plt.figure(figsize = (25,25))
# Plot each year's time series in its own facet
g = sns.relplot(
    data=grouped,
    x="year", y='Sales', col="Sub-Category", hue="Ship Mode",
    kind="line", palette="crest", linewidth=4, zorder=5,
    col_wrap=3, height=2, aspect=2, legend=False,
)
# Iterate over each subplot to customize further
for year, ax in g.axes_dict.items():

    # Add the title as an annotation within the plot
    ax.text(.8, .85, year, transform=ax.transAxes, fontweight="bold")

    # Plot every year's time series in the background
    sns.lineplot(
        data=grouped, x="year", y="Sales", units="year",
        estimator=None, color=".7", linewidth=1, ax=ax,
    )

# Reduce the frequency of the x axis ticks
#ax.set_xticks(ax.get_xticks()[::1])

# Tweak the supporting aspects of the plot
g.set_titles("")
g.set_axis_labels("", "Sales")
g.tight_layout()
```

```python
sns.set_theme(style="dark")
plt.figure(figsize = (25,25))
# Plot each year's time series in its own facet
g = sns.relplot(
    data=grouped,
    x="year", y='Quantity', col="Sub-Category", hue="Ship Mode",
    kind="line", palette="pastel", linewidth=4, zorder=5,
    col_wrap=3, height=2, aspect=2, legend=False,
)
# Iterate over each subplot to customize further
for year, ax in g.axes_dict.items():

    # Add the title as an annotation within the plot
    ax.text(.8, .85, year, transform=ax.transAxes, fontweight="bold")

    # Plot every year's time series in the background
    sns.lineplot(
        data=grouped, x="year", y="Quantity", units="year",
        estimator=None, color=".7", linewidth=1, ax=ax,
    )

# Reduce the frequency of the x axis ticks
#ax.set_xticks(ax.get_xticks()[::1])

# Tweak the supporting aspects of the plot
g.set_titles("")
g.set_axis_labels("", "Quantity")
g.tight_layout()
```
